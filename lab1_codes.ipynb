## Outlinks data - preparing outlinks dataframe to save to AWS RDS

```python
##creating outlinks table data
outlinks=df[['Tweet Id','Outlinks']]
outlinks=outlinks.assign(Outlinks=outlinks.Outlinks.str.split(",")).explode('Outlinks')
outlinks['Outlinks'] = outlinks['Outlinks'].str.replace(r'[\[\]]', '',regex=True)
outlinks['Outlinks']=outlinks['Outlinks'].str.replace("'", "")
outlinks.columns=['tweet_id','outlinks']
outlinks.outlinks = outlinks.outlinks.str.strip()
outlinks = outlinks.replace('', pd.NA)
outlinks = outlinks.dropna()
outlinks = outlinks.drop_duplicates()

## Removing 'http://Pictory.ai' from outlinks data as this is a duplicate with different case for same tweet id
outlinks = outlinks.loc[~((outlinks['outlinks'] == 'http://Pictory.ai') & (outlinks['tweet_id'] == 1617402821162270720))]
```

```python
outlinks.isna().sum()
```

## CAUTION - THIS SAVES DATA TO AWS MYSQL

```python
outlinks.to_sql('outlinks', con=engine, if_exists='append', index=False)
```

## Hashtag data - preparing hashtag dataframe to save to AWS db

```python
##creating hashtag table data
hashtag=df[['Tweet Id','hashtag']]
hashtag=hashtag.assign(hashtag=hashtag.hashtag.str.split(",")).explode('hashtag').reset_index(drop=True)
hashtag['hashtag'] = hashtag['hashtag'].str.replace(r'[\[\]]', '',regex=True)
hashtag['hashtag']=hashtag['hashtag'].str.replace("'", "")
hashtag.columns=['tweet_id','hashtag']
hashtag.hashtag = hashtag.hashtag.str.strip()
hashtag = hashtag.replace('', pd.NA)
hashtag.hashtag = hashtag.hashtag.str.lower()
hashtag = hashtag.dropna()
hashtag = hashtag.drop_duplicates()
```

## CAUTION - THIS SAVES DATA TO AWS MYSQL

```python
hashtag.to_sql('hashtag', con=engine, if_exists='append', index=False)
```
